```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('householdtask3.csv')
```

```python
data.head(10)
```

```python
# Line chart

plt.plot(data['age'])
plt.plot(data['size'])

# Title
plt.title("LINE CHART")

# x label and y label 
plt.xlabel('age')
plt.ylabel('size')

#Legend
plt.legend(["age","size"])


# Result
plt.show()
```

```python
# Bar chart

plt.bar(data['age'], data['size'])

# Title
plt.title("BAR CHART")

# x label and y label
plt.xlabel('age')               
plt.ylabel('size')

# Result
plt.show()
```
